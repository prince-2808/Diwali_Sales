```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali.csv', encoding='unicode_escape')
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
#>>>>> drop columns
df.drop(['Status','unnamed1'] ,axis=1  , inplace=True ) # del two col
```

```python
df.isnull().sum() # chek total null value
```

```python
#drop null value in df data
df.dropna(inplace= True) # inplace = true save your drop
```

```python
df.isnull().sum() # cheking null value
```

```python
#>>>> change data type of column
#if you use df.info then see (Amount) is float data type but i dont want decimal

df['Amount']=df['Amount'].astype('int')

df['Amount'].dtype # chek datatype
```

```python
#column names
df.columns
```

```python
# rename a column
df.rename( columns={ 'Marital_Status': 'Merrage'  }   )
```

```python
# describe used to chek (count, mean ,max every aggr)
df['Amount'].describe()
```

```python
# >>>>>>
# EDA- expolatery data analysis
# Q.>i want to see total male and female on bar chart

ax= sns.countplot( x='Gender' , data= df  ) #showing bar plot

for bar in ax.containers:   # it show leble 7832 and 3407
    ax.bar_label(bar)
```

```python
#total amount of purchase by male and female

group_gen=df.groupby('Gender'  ).agg({'Amount': 'sum'} )
group_gen
```

```python
mf = sns.barplot( x='Gender'    , y='Amount'     , data= group_gen   )


for bar in mf.containers: # show label
    mf.bar_label(bar)
```

For the above graphes we can see the most of the buyers are female and the purchasing power female is greater than male

```python
# count of buyers by age group and gender

ax= sns.countplot( x= 'Age Group' , hue='Gender', data=df) # hue is legends
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
# >>>>>>>>>>>>>>>>>>>>>
# >>>>>>>>>>>>>>>>>>>>>
# age group wise total buying amount

grp= df.groupby('Age Group').agg({'Amount': 'sum'  })
grp
```

```python
age_amount=sns.barplot(x= 'Age Group', y='Amount', data= grp  )


for bar in age_amount.containers:
    age_amount.bar_label(bar)
    
```

From the above graphs we can see that most the buyers are the age group brtween 26-35 yr females

STATE


```python
#total number of orders from top 10 states

grp= df.groupby('State').agg( { 'Orders': 'sum' } ). sort_values(by='Orders', ascending= False).head(10)
grp
```

```python
sta_or=sns.barplot(x='State', y='Orders',data= grp)
for bar in sta_or.containers:
     sta_or.bar_label(bar)   
    
#increase graph size
sns.set( rc={ 'figure.figsize' :(17,5) } )
```

```python
# total amount of sales from top 10 states

grp=df.groupby('State').agg({'Amount': 'sum'}).sort_values(by='Amount', ascending= False).head(10)
grp
```

```python
sns.barplot(x='State', y='Amount',data=grp)
```

from the above graphs we can see that most of the order / and sales amount from UP,MH,karnataka
